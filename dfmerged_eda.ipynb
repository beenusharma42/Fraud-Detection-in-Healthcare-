```python
df1['ClaimStartDt']=pd.to_datetime(df1['ClaimStartDt'])
df1['ClaimEndDt']=pd.to_datetime(df1['ClaimEndDt'])
df1['AdmissionDt']=pd.to_datetime(df1['AdmissionDt'])
df1['DischargeDt']=pd.to_datetime(df1['DischargeDt'])
df1['DOB']=pd.to_datetime(df1['DOB'])
df1['DOD']=pd.to_datetime(df1['DOD'])
df1.dtypes                                          #Changing the datatype of columns containing dates to datetime format
```

Finding the null values in each column:

```python
df1.isnull().sum()             #returns the number of null values present column-wise
```

```python
null_values = df1.isnull().sum().sort_values(ascending=False)
Percent_NUll = ((df1.isnull().sum()/df1.shape[0])*100)
Percent_NUll = Percent_NUll.sort_values(ascending=False)
Null_data = pd.concat([null_values,Percent_NUll], axis=1, keys=['null_values', 'Percent_NUll'])
Null_data
```

Dropping column 'ClmProcedureCode_6' since it has 100percent Null values

```python
df1.drop(['ClmProcedureCode_6'], axis = 1, inplace = True)
```

#### Null value imputation for quantitative data:

```python
num_columns= df1.select_dtypes(include=np.number)
num_col= num_columns.columns
numcol_list= list(num_col)
numcol_list
```

```python
df1.update(df1[numcol_list].fillna(0))
```

```python
df1[numcol_list].isnull().sum()
```

#### Null value imputation for categorical data:

```python
catg_columns=df1.select_dtypes(include=object)
catg_col=catg_columns.columns
catgcol_list=list(catg_col)
catgcol_list
```

Creating a new category 'NA' for missing categorical data

```python
df1.update(df1[catgcol_list].fillna('NA'))
df1[catgcol_list].isnull().sum()
```

Null value imputation for date columns:

```python
date_columns=df1.select_dtypes(include= ['datetime64[ns]'])
date_col= date_columns.columns
datecol_list=list(date_col)
datecol_list
```

```python
df1[datecol_list].isnull().sum()
```
